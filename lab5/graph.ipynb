```python
from matplotlib import pyplot as plt
import pandas as pd
from matplotlib import style
import matplotlib.ticker as ticker
import numpy as np

df_1000 = pd.read_csv('./data/data_python.csv', header= None)
df_1000 = df_1000.sort_values(0)
df_1000 = df_1000.groupby(np.arange(len(df_1000))//10).mean()
pd.set_option('display.max_rows', 2000)
df_1000.columns = ['Capacity of the Knapsack','Time in milliseconds']
print(df_1000)
x1 = df_1000.iloc[:, 0]
y1 = df_1000.iloc[:, 1]*1000
m1, c1 = np.polyfit(x1, y1, 1)
#add fitted regression equation to plot
equation_1 = 'Line of best fit: y = ' + '{:.2f}'.format(c1) + ' + {:.2f}'.format(m1) + '*x'

fig = plt.figure(figsize=(10,5), dpi = 500)
plt.xlabel("Capacity of the Knapsack")
plt.ylabel("Average time taken in milliseconds")
plt.title("Average time taken to find a solution for 1000 items with different knapsack capacities")
plt.scatter(x1, y1, color='#00FFFF')
plt.plot(x1, m1*x1+c1, color='#00FFFF', linestyle='--', linewidth=2)
plt.xticks([200, 400, 600, 800, 1000, 1200, 1400, 1600, 1800, 2000])
plt.legend(['Data points of average time taken with different knapsack capacity', equation_1])
plt.show()
fig.savefig('./Graph_Pic/1000.png', dpi= 500)

```
